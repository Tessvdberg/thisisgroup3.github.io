## Paralympics
Paris is hosting the Paralympics in 2024. One of the events at the Paralympics is an open water swimming event in the Seine (apparently the water is clean or so the French say). Since the City of Amsterdam thinks it is better than Paris, they want to host an event before the Paralympics, snubbing the Parisians. The idea is to host a 5km. open water swimming event through the canals of Amsterdam. You are asked by the municipality of Amsterdam to advise on the feasibility of the event from the perspective of the safety of the partaking athletes from an environmental perspective. The event is going to be hosted in May.

#### Questions about Canal Boats.

#### Answers.  

 We are starting with the existing distribution of canal boats, by examine the current division of diesel/fossil fuel-driven boats and electric ones. According to the ‘new’ policy of Amsterdam’s authorities, in 2025 all diesel/fossil fuel engines need to be banned from the waters (Sterling, 2020). As stated by Sterling (2020), a substantial proportion of all commercial vessels, specifically 75% of the total 550, have achieved emissions-free status. The 320 canal boats were fit to go first, as they constitute one of the most popular tourist attractions in Amsterdam. 

 The frequency and duration of boat usage throughout the day play a crucial role in understanding the feasibility of organizing the event safely. During summer season, canal boats make up approximately 24% of the total boats on Amsterdam’s waterways (Van den Bogaard et al., 2013). The figure below shows the amount of canal boats crossing the Prinsengracht every hour a day, in summer, spring and fall season. Following timeframes in spring (may) during 09:00 – 12:00 &  13:00 – 16:00, the density of canal boats crossing the Prinsengracht canal is relatively lower compared to other times of the day (Van den Bogaard et al., 2013).   

Van Mensch et al. (2016) argue that the total demand would be ca. 9105 MWh, given that all vessels across various segments were fully electric and operated for the same number of hours annually. This energy demand could be seen as an opportunity for energy providers, since they can make 0,12€/kWh and about 1.1 M€ in total. To get a better understanding of how this energy use refers back to the household, a typical household consumes around ca. 3500 kWh of electricity (Van Mensch et al., 2016). This means if all vessels were transitioned to electric driven / emission free propulsion, the energy demand would be equivalent to that of approximately 2,600 households, making up less then 0,01% of our national electricity consumption.

Due to the regular usage of canal and rental boats, the transition would probably only be beneficial for these specific vessels. As we delve deeper into this topic, it raises a significant social and economic question regarding who should bear the financial burden of this transition, especially in the private boat sector where boat ownership is often considered a luxury.  

In addition, the availability of charging stations is a critical factor, which needs to be overseen by either the municipality of Amsterdam or Waternet. Currently, the municipality is conducting research to determine the necessary number of charging stations, estimated to range between 1,000 and 1,500 stations.    

In the context of organizing the swimming event, it's essential to consider that, when relying solely on electric vessels, the chosen route should be in proximity to existing charging stations.

#### Questions about Clean Energy during the Event.  
- How many support boats and vehicles are needed for the Paralympics event only?
- If only clean energy can be used, how many solar panels or wind turbines are needed? 
- Is it possible to accommodate these facilities within the city? 
- Would their be any effect on the water quality if there are less/no canal boats using fossil fules? 

#### Answers. 

In their document on the “World Para Swimming Rules and Regulations”, the IPC does not make any concrete statements on the safety during swim events. They only state that participants are responsible for their own medical support (SWANS, 2018). Therefore, we compare the swim event to the yearly held Amsterdam city swim. The event has 3000 participants and safety measures are provided by the KNMR. During the event, 40 lifeguards, 3 waterjets, one rescue boat and a crew of 26 people that is normally stationed at KNRM Huizen-Blaricum (City Swim, 2023). Besides, every 500m there is a resting point and a first aid station (City Swim, 2023). The city swim is in total 2,1 km long, while the Paralympics event is supposed to be 5 km long (City Swim, 2023). Therefore, assuming that the number of participants is similar, we would double the safety measures of the city swim. Thus, the event needs 2 rescue boats and 6 waterjets. 

We expect the rescue boats to have an average energy use of 15,6 kWh and the waterjet an energy use of 25 kWh (Van Mensch et al., 2016; Olsen, 2008). If both the wind turbines and the solar energy systems operate on a 4KWh power rate, the wind turbine produces 21.6 kWh and the solar system 15,6 kWh of energy (Edwards, 2023). Consequently, the event needs 6,9 wind turbines or 9,6 solar panel systems. However, here it is important to note that this calculation assumes that the vehicles are running for the entire time. Thus, in practice less might be necessary. 

In their guidelines for sustainable event planning, the municipality of Amsterdam argues that events should consume 100% green energy (Gemeente Amsterdam, 2023). The municipality has a map with all energy points that can be used for events throughout the city (Gemeente Amsterdam, 2023). These electricity points are providing green energy. Therefore, the swim event should be located at one of these points to be able accommodate all facilities (see map below).

What impact can a change in boat energy use have on Amsterdam? During the Covid pandemic, the canals of Venice got notably clearer (Clifford, 2020). This can be traced back to the decrease of boat traffic due to the pandemic (Clifford, 2020). The clearer water though occurred because sediments stayed at the bottom of the canals and not because the water became cleaner (Clifford, 2020). What was noticed though is that the air quality increased significantly (Clifford, 2020). Therefore, a switch to less canal boats using fossil fuels in Amsterdam might not necessarily led to a better water quality, but a better air quality. However, fossil fuel boats can lead to chemical pollution of the canal through leaking fuel etc. (Sagerman et al. 2019: p. 518). This does have an impact on the water quality and can be prohibited by decreasing boat traffic and switching to electric boats (Sagerman et al., 2019, p. 518).
